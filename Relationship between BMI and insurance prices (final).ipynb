# U.S. Medical Insurance Costs

The analysis of the health care data reveals notable patterns in insurance charges based on BMI categories. Individuals with a BMI in the healthy range (0 to 24.9) tend to incur lower overall charges, with an average difference of 3683.34 USD compared to those with a BMI above 24.9. The highest charge observed in the healthy BMI group is 33,947.50 USD, contrasting significantly with the maximum charge of 63638.92 USD for individuals with a BMI above 24.9. On average, the healthy BMI group pays less than the unhealthy BMI group. However, it is essential to note that while a probability exists that BMI correlates with insurance charges, the data does not exhibit a strong enough correlation to confidently assess the impact of BMI on prices.
```python
rounded_bmis = []
other_data = []

with open('insurance.csv', 'r') as insurance_data:
    csv_reader = csv.reader(insurance_data)
    
    header = next(csv_reader, None)
    
    for row in csv_reader:
        bmi = float(row[2])
        rounded_bmi = round(bmi, 1)
        rounded_bmis.append(rounded_bmi)
        
        other_data.append(row[:2] + [rounded_bmi] + row[3:])
        
for rounded_bmi, other_values in zip(rounded_bmis, other_data):
    print(f"Rounded BMI: {rounded_bmi}, Other Data: {other_values}")

```

```python
# Create empty lists to store 'bmi' and 'charges' values
bmi_values = []
charges_values = []

with open('insurance.csv', 'r') as insurance_data:
    csv_reader = csv.reader(insurance_data)
    
    # Skip the header row if it exists
    header = next(csv_reader, None)
    
    for row in csv_reader:
        # Assuming 'bmi' is the column index for BMI and 'prices' for charges
        bmi = float(row[2])  # Convert BMI to float
        charge = float(row[6])  # Convert prices to float
        
        bmi_values.append(bmi)
        charges_values.append(charge)

# Print the 'bmi' and 'charges' values
for bmi_val, charge_val in zip(bmi_values, charges_values):
    print(f"BMI: {bmi_val}, Charges: {charge_val}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.regplot(x=bmi_values, y=charges_values)
plt.title('Regression Plot of BMI vs. Charges')
plt.xlabel('BMI')
plt.ylabel('Charges')
plt.show()
```

```python
import csv

# Create empty lists to store 'bmi' and 'charges' values
bmi_values = []
charges_values = []

with open('insurance.csv', 'r') as insurance_data:
    csv_reader = csv.reader(insurance_data)
    
    # Skip the header row if it exists
    header = next(csv_reader, None)
    
    # Initialize variables for counting and calculating averages and means
    total_healthy_bmi = 0
    total_healthy_bmi_charges = 0
    total_above_24_9 = 0
    total_above_24_9_charges = 0
    total_individuals = 0
    total_charges = 0
    
    # Initialize variables for range calculation
    healthy_bmi_charges_list = []
    above_24_9_charges_list = []
    
    for row in csv_reader:
        # Assuming 'bmi' is the column index for BMI and 'prices' for charges
        bmi = float(row[2])  # Convert BMI to float
        charge = float(row[6])  # Convert prices to float
        
        bmi_values.append(bmi)
        charges_values.append(charge)
        
        # Classify individuals based on BMI
        if bmi <= 24.9:
            total_healthy_bmi += 1
            total_healthy_bmi_charges += charge
            healthy_bmi_charges_list.append(charge)
        else:
            total_above_24_9 += 1
            total_above_24_9_charges += charge
            above_24_9_charges_list.append(charge)
        
        # Accumulate totals for mean calculation
        total_individuals += 1
        total_charges += charge

# Calculate averages
average_healthy_bmi_charge = total_healthy_bmi_charges / total_healthy_bmi if total_healthy_bmi != 0 else 0
average_above_24_9_charge = total_above_24_9_charges / total_above_24_9 if total_above_24_9 != 0 else 0

# Calculate means
mean_charge = total_charges / total_individuals if total_individuals != 0 else 0
mean_healthy_bmi_charge = sum(healthy_bmi_charges_list) / len(healthy_bmi_charges_list) if len(healthy_bmi_charges_list) != 0 else 0
mean_above_24_9_charge = sum(above_24_9_charges_list) / len(above_24_9_charges_list) if len(above_24_9_charges_list) != 0 else 0

# Calculate ranges
range_healthy_bmi = max(healthy_bmi_charges_list) - min(healthy_bmi_charges_list) if len(healthy_bmi_charges_list) > 1 else 0
range_above_24_9 = max(above_24_9_charges_list) - min(above_24_9_charges_list) if len(above_24_9_charges_list) > 1 else 0

# Print results
print(f"Total individuals with healthy BMI (0 to 24.9): {total_healthy_bmi}")
print(f"Average charge per individual with healthy BMI: {average_healthy_bmi_charge:.2f}")
print(f"Mean charge per individual with healthy BMI: {mean_healthy_bmi_charge:.2f}")
print(f"Range of charges for individuals with healthy BMI: {range_healthy_bmi:.2f}")

print(f"\nTotal individuals with BMI above 24.9: {total_above_24_9}")
print(f"Average charge per individual above 24.9: {average_above_24_9_charge:.2f}")
print(f"Mean charge per individual above 24.9: {mean_above_24_9_charge:.2f}")
print(f"Range of charges for individuals above 24.9 BMI: {range_above_24_9:.2f}")

print(f"\nTotal individuals in the dataset: {total_individuals}")
print(f"Mean charge per individual: {mean_charge:.2f}")
```
