Introduction: This study aims to delve into a dataset encompassing the GDP and life expectancy data of four countries across time. Through exploratory analysis and visualizations, we seek to uncover and understand the relationship between life expectancy and GDP, as well as the trends in life expectancy over time.

Summary: This study examines the relationship between GDP and life expectancy across four countries over time. Initial analysis reveals Germany consistently leading in life expectancy, with China at the lowest. Despite Germany's lead, there's a general upward trend in life expectancy across all countries. Interestingly, China's rapid GDP surge doesn't directly align with a proportional increase in life expectancy, while Chile, with the lowest GDP, shows unexpectedly high life expectancy. This suggests a positive relationship between life expectancy and GDP, but it's not definitive: lower GDP doesn't always mean lower life expectancy, and a rising GDP may indicate an upward trend in life expectancy. Factors beyond GDP, such as pandemics and wars, may significantly influence life expectancy.

```python
germany_data = data[data['Country'] == 'Germany']

china_data = data[data['Country'] == 'China']

chile_data = data[data['Country'] == 'Chile']

mexico_data = data[data['Country'] == 'Mexico']

plt.figure(figsize=(10,5))

# Subplot 1
plt.subplot(1, 4, 1)
plt.boxplot(germany_data['Life expectancy at birth (years)'], boxprops=dict(color='green'))
plt.title('Germany')
plt.ylabel('Life Expectancy in Years')
plt.ylim(71, 82)

# Subplot 2
plt.subplot(1, 4, 2)
plt.boxplot(china_data['Life expectancy at birth (years)'], boxprops=dict(color='purple'))
plt.title('China')
plt.ylabel('Life Expectancy in Years')
plt.ylim(71, 82)

# Subplot 3
plt.subplot(1, 4, 3)
plt.boxplot(chile_data['Life expectancy at birth (years)'], boxprops=dict(color='blue'))
plt.title('Chile')
plt.ylabel('Life Expectancy in Years')
plt.ylim(71, 82)

# Subplot 4
plt.subplot(1, 4, 4)
plt.boxplot(mexico_data['Life expectancy at birth (years)'], boxprops=dict(color='red'))
plt.title('Mexico')
plt.ylabel('Life Expectancy in Years')
plt.ylim(71, 82)

plt.subplots_adjust(wspace=1.0)
plt.show()
```

Above, the box plot showcases the distribution of life expectancy across the years for each country within the dataset. Notably, Germany boasts the highest mean, while China demonstrates the lowest. Additionally, Mexico presents the smallest range, in contrast to China with the widest range. To further explore these variations, let's employ a line chart to track the changes in life expectancy for each country over time.

```python
countries_of_interest_line = ['Germany', 'China', 'Chile', 'Mexico']
selected_countries_line_data = data[data['Country'].isin(countries_of_interest_line)]

plt.figure(figsize=(10, 6))
sns.lineplot(data=selected_countries_line_data, x='Year', y='Life expectancy at birth (years)', hue='Country')
plt.title('Life Expectancy Over Time')
plt.xlabel('Year')
plt.ylabel('Life Expectancy in Years')
```

The line graph above vividly illustrates a consistent positive correlation between the progression of years and life expectancy. Notably, Germany consistently maintains the highest life expectancy annually, followed by Chile, Mexico, and China in descending order. This prompts the question of why Germany consistently leads in life expectancy across the years, as well as why there's a general upward trend in life expectancy across all countries in the dataset.

As the dataset solely provides GDP for comparison, our exploration pivots to understanding the factors influencing these trends. Our focus shifts towards comparing the life expectancy of these four countries against their GDP. This comparative analysis seeks to uncover whether an increase in GDP is associated with higher life expectancy, shedding light on potential correlations.

```python
countries_of_interest_line = ['Germany', 'China', 'Chile', 'Mexico']
selected_countries_line_data = data[data['Country'].isin(countries_of_interest_line)]

plt.figure(figsize=(10, 6))
sns.lineplot(data=selected_countries_line_data, x='Year', y='GDP', hue='Country')
plt.title('GDP Over Time')
plt.xlabel('Year')
plt.ylabel('GDP')
```

The presented graph illustrates the GDP trends for each country over time. Notably, China emerges as a frontrunner post-2006, while Chile and Mexico maintain relatively stable GDP trajectories throughout the dataset's timeline. A compelling observation arises when correlating China's GDP surge with its upward trend in the life expectancy graph. This alignment sparks curiosity as to why their economic growth surpassed the corresponding increase in life expectancy.

Equally intriguing is Chile's consistent position with the lowest GDP among the countries. Despite this, their life expectancy has not only increased along with the others but has also ranked second among the countries in this dataset. This divergence prompts us to consider factors beyond GDP that influence a country's life expectancy.

While it appears that GDP holds some influence on life expectancy trends, these findings suggest that there are other significant contributors impacting a country's life expectancy beyond economic measures.

Is there a correlation betwen GDP and life expectancy? 

```python
countries_of_interest = ['China']
selected_countries_data = data[data['Country'].isin(countries_of_interest)]

plt.figure(figsize=(8,6))
x_values = selected_countries_data['Life expectancy at birth (years)']
y_values = selected_countries_data['GDP']
plt.scatter(x_values, y_values, linestyle='-')
plt.xlabel('Life Expectancy in Years')
plt.ylabel('GDP')
plt.title('Life Expectancy vs GDP :China')

countries_of_interest_2 = ['Chile']
selected_countries_data_2 = data[data['Country'].isin(countries_of_interest_2)]

plt.figure(figsize=(8,6))
x_values = selected_countries_data_2['Life expectancy at birth (years)']
y_values = selected_countries_data_2['GDP']
plt.scatter(x_values, y_values, linestyle='-')
plt.xlabel('Life Expectancy in Years')
plt.ylabel('GDP')
plt.title('Life Expectancy vs GDP :Chile')

countries_of_interest_3 = ['Germany']
selected_countries_data_3 = data[data['Country'].isin(countries_of_interest_3)]

plt.figure(figsize=(8,6))
x_values = selected_countries_data_3['Life expectancy at birth (years)']
y_values = selected_countries_data_3['GDP']
plt.scatter(x_values, y_values, linestyle='-')
plt.xlabel('Life Expectancy in Years')
plt.ylabel('GDP')
plt.title('Life Expectancy vs GDP :Germany')

countries_of_interest_4 = ['Mexico']
selected_countries_data_4 = data[data['Country'].isin(countries_of_interest_4)]

plt.figure(figsize=(8,6))
x_values = selected_countries_data_4['Life expectancy at birth (years)']
y_values = selected_countries_data_4['GDP']
plt.scatter(x_values, y_values, linestyle='-')
plt.xlabel('Life Expectancy in Years')
plt.ylabel('GDP')
plt.title('Life Expectancy vs GDP :Mexico')
```

Conclusion: While observing the life expectancy versus GDP graphs, there's a temptation to conclude a direct relationship: as GDP rises, so does life expectancy. Undoubtedly, there exists a discernible correlation between the two factors. However, a closer examination reveals instances where countries with lower GDP exhibit higher life expectancy than those with higher economic measures. It might be more accurate to consider a positive relationship between life expectancy and GDP, indicating that as life expectancy rises, GDP tends to follow suit. Yet, it's crucial to note that a high GDP doesn't guarantee a high life expectancy, but a rising GDP could potentially signal an increasing life expectancy trend. Naturally, numerous other factors play vital roles here; for instance, pandemics and wars can significantly impact life expectancy, potentially reducing it.
